## Zajem podatkov

```python
import pandas as pd

%matplotlib inline
igralci=pd.read_csv('igralci.csv', index_col='ID')

igralci
```

#### LEGENDA

G: Koliko tekem je igralec oddigral v sezoni <br>
GS: Kolikokrat je bil igrale v prvi petorki <br>
Min: Koliko minut igra igralec v povprečju na tekmi <br>
Tocke: Koliko točk da igralec v povprečju na tekmo <br>
2FG: V procentih koliko košev za dve piki igralcev v povprečju zadane na tekmo <br>
3FG: V procentih koliko trojk igralec v povprečju zadane na tekmo <br>
FT: V procentih koliko prostih metov igralec v povprečju zadane na tekmo <br>

Ker python ni sprejel številk v imenih stolpcev lahko zdaj zamenjamo.

```python
igralci.rename(columns={'DvaFG':'2FG', 'TriFG':'3FG'}, inplace=True)
```

Razvrstimo igralce po velikosti.

```python
igralci.sort_values('Visina')
```

## Osnovna analiza

Najprej lahko pogledamo kakšne nacionalnosti so igralci.

```python
igralci.groupby('Nacionalnost').count()['Ime'].sort_values()
```

```python
igralci.groupby('Nacionalnost').count()['Ime'].sort_values().plot(kind='pie', figsize=(15, 15))
```

Zanimivo je to, da je največ igralcev iz ZDA, glede na to, da imajo tam posebno košarkaško ligo imenovano NBA. To pomeni, da je v povprečju med dobrimi košarkaši več Američanov kot Evropejcev.

Sledita histograma porazdelitve igralcev po višini in po starosti.

```python
igralci['Visina'].plot(kind = 'hist').set(xlabel = 'Višina', ylabel = 'Število igralcev')
```

```python
igralci['Letnica_rojstva'].plot(kind = 'hist').set(xlabel = 'Letnica rojstva', ylabel = 'Število igralcev')
```

Naredimo nov stolpec, ki bo vseboval starost igralcev.

```python
igralci['Starost'] = 2018 - igralci['Letnica_rojstva']
igralci['Starost']
```

Starost ni natančna za vsakega igralca, saj se gleda le po letih in ne po točnem datumu!

## Analiza višine igralcev

Za boljšo nadaljno analizo pretvorimo višino na 5cm.

```python
igralci['Pet_cm'] = 0.05*(igralci['Visina']//0.05)
igralci
```

Izračunajmo različna povprečja za skupino 5cm.

```python
igralci.groupby('Pet_cm').mean()[['G', 'GS', 'Tocke', '2FG', '3FG', 'FT']]
```

Zanimivo je to, da v povprečju manjši igralci dosežejo več točk, čeprav igrajo na manj tekmah kot višji igralci.
Poglejmo si še graf vvišine v odvisnosti od točk, na katerem lahko potrdimo zgornjo trditev.
Opazimo lahko tudi veliko razliko v povprečnem procentu zadetih trojk, torej lahko rečemo, da igralci, manjši od 2m, zadanejo več trojk na tekmo, čeprav je 33% relativno slabo. <br>
Poglejmo si, če je razmerje med igralci višjih od 2m in nižjih od 2m približno enako.

```python
igralci['Pet_cm'].value_counts()
```

```python
igralci.sort_values('Pet_cm', ascending = True).plot(x='Pet_cm',y='Tocke', kind = 'scatter')
```

Iz grafa lahko preberemo, da igralci, ki spadajo v skupino 1.9 cm, dajo več točk od ostalih.

```python
igralci.sort_values('Pet_cm', ascending = True).plot(x='Pet_cm',y='3FG', kind = 'scatter')
```

Pri trojkah se pri vseh skupinah povpreje drži na relativno enaki ravni, čeprav lahko opazimo nekatera odstopanja.

Upoštevati moramo tudi take primere, kjer so eni igralci malo igrali in niso metali na koš oz. niso metali trojk, zato imajo 0%, kar pa vpliva na skupno povprečje.

## Analiza starosti igralcev

```python
igralci.sort_values('Starost').tail()
```

Zgornja tabela prikazuje 5 najstarejših igralcev. Poglejmo še 5 najmlajših:

```python
igralci.sort_values('Starost').head()
```

Najmlajši igralec je naš košarkaš Luka Dončič. <br>
Poglejmo si, kako leta vplivajo na statistiko:

```python
igralci.groupby('Starost').mean()[['G', 'GS', 'Tocke', '2FG', '3FG', 'FT']]
```

Že takoj lahko vidimo, da imaš Luka Dončič pri 19-ih letih tudi vrhunsko statistiko, ki je nad povprečna.

```python
igralci.sort_values('Starost', ascending = True).plot(x='Starost',y='Min', kind = 'scatter')
```

Na zgornjem grafu lahko opazimo, da v povprečju dobijo več minut starejši igralci, kar je tudi pričakovano, saj bi lahko rekli, da imajo več izkušenj.

```python
igralci.groupby('Starost').mean()[['G', 'GS', 'Tocke', '2FG', '3FG', 'FT']].plot(figsize=(12.5, 8), xlim=(19, 41))
```

```python
igralci.plot.scatter(x='Starost', y='Visina', s=igralci['Tocke']*20, figsize=(15, 10));
```

Na zgornjem grafku je prikazano razmerje med starostjo, višino in povprečnimi točkami na tekmo. Večja kot je točka, več točk je dal igralec na tekmo.

## Analiza klubov

Najprej si poglejmo koliko igralcev ima določen klub.

```python
igralci.groupby('Klub').count()['Ime'].sort_values()
```

Zdaj pa si poglejmo povprečno statistiko za vsak klub.

```python
igralci.groupby('Klub').mean()[['G', 'GS', 'Tocke', '2FG', '3FG', 'FT']].plot.bar(figsize=(17.5, 8.5))
```

Lahko opazimo večje razlike pri prostih metih in trojkah, zato si poglejmo še bolj nazorno tabelo.

```python
igralci.groupby('Klub').mean()
```

Iz tabele izberimo prve štiri klube glede na končne rezultate. Informacije dobimo na spletni strani.

```python
 najboljsi = igralci[
    (igralci.Klub =='Real Madrid') | 
    (igralci.Klub =='CSKA Moscow') | 
    (igralci.Klub =='Olympiacos Piraeus') | 
    (igralci.Klub =='Panathinaikos Superfoods Athens')
]
najboljsi[['Klub', 'Tocke', 'FT', '3FG', '2FG']].boxplot(by='Klub', figsize=(12,11), rot=45)
```


1. Real Madrid <br>
2. CSKA Moscow <br>
3. Olympiacos Piraeus <br>
4. Panathinaikos Superfoods Athens

Čeprav je lani Real Madrid zmagal, mu tesno sledi CSKA Moscow in ima pri trojkah ter prostih metih boljšo porazdelitev. Res pa je, da so njihove najslabše vrednosti nižje kot pri zmagovalcih. Podobna primerjava velja tudi pri ostalih dveh klubih.

Lahko bi zaključili analizo tako, da na končen rezultat ne vplivajo vedno višina, starost ali nacionalnost, ampak je potrebno poleg talenta vložiti tudi veliko dela in truda. Lahko smo ponosni na naše košarkarje, ki so letos dosegli veliko zmag in nekateri so uspešni tudi v tujini.
