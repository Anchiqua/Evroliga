## Zajem podatkov

Razvrstimo igralce po velikosti.

## Osnovna analiza

Najprej lahko pogledamo kakšne nacionalnosti so igralci.

```python
igralci.groupby('Nacionalnost').count()['Ime'].sort_values()
```

Presenetljivo je to, da je največ igralcev nacionalnosti ZDA, saj imajo tam posebno ligo imenovano NBA. Lahko pridemo do ugotovitve, da je v ZDA preveč dobrih igralcev, ki bi lahko igrali le v NBA. Opazimo lahko še to, da imamo 6 slovenskih košarkarjev v evroligi, kar je tudi dober dosežek. 

Zdaj pa si oglejmo še koliko igralcev ima določen klub.

```python
igralci.groupby('Klub').count()['Ime'].sort_values()
```

Sledita histograma porazdelitve igralcev po višini in po starosti.

```python
igralci['Visina'].plot(kind = 'hist').set(xlabel = 'Višina', ylabel = 'Število igralcev')
```

```python
igralci['Letnica_rojstva'].plot(kind = 'hist').set(xlabel = 'Višina', ylabel = 'Število igralcev')
```

## Analiza višine igralcev

Za boljšo nadaljno analizo pretvorimo višino na 5cm.

```python
igralci['5cm'] = 0.05*(igralci['Visina']//0.05)
igralci
```

Izračunajmo različna povprečja za skupino 5cm.

```python
igralci.groupby('5cm').mean()[['GS' ,'G','Tocke']]
```

Zanimivo je to, da v povprečju najmanjši igralci dosežejo več točk.
